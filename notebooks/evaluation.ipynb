```python
import pandas as pd
import matplotlib.pyplot as plt
import os
import sys
import warnings
warnings.filterwarnings('ignore')
dir_store = os.path.join("..", "data")
sys.path.append(os.path.join( ".."))
from src.utils.store_utils import StorageHandler
store_handler = StorageHandler(dir_store=dir_store)
```

```python
X_test_Legit = store_handler.read("X_test_Legit.p")
X_test_Fraud = store_handler.read("X_test_Fraud.p")
pred_test_Legit = store_handler.read("pred_test_Legit.p")
pred_test_Fraud = store_handler.read("pred_test_Fraud.p")
```

```python
import seaborn as sns

import matplotlib.pyplot as plt
```

```python
fig = plt.figure(figsize=(6, 5))
fig.set_size_inches(23.5, 23.5)
c = 0
for column in range(30):
    plt.subplot(6, 5, c + 1)
    c += 1

    sns.distplot(X_test_Legit[:,column]-pred_test_Legit[:,column], hist=True, kde=True, 
             bins=100, color = 'darkblue', 
             hist_kws={'edgecolor':'black'},
             kde_kws={'linewidth': 4})

    plt.grid()
plt.legend(loc='upper center', bbox_to_anchor = (0.5, -0.05),fancybox = True, shadow = True)
plt.show()
```
